```python
import random

def menu ():
    print ('''Menu:
    (1) Criar Lista 
    (2) Ler Lista
    (3) Soma
    (4) Média
    (5) Maior
    (6) Menor
    (7) estaOrdenada por ordem crescente
    (8) estaOrdenada por ordem decrescente
    (9) Procura um elemento
    (0) Sair''')
    return

def criaListaR (N):
    listaR = []
    while (N > 0):
        num = random.randrange(101)
        listaR.append(num)
        N = N - 1
    return listaR

def criaLista (N):
    listaN = []
    while (N > 0):
        num = int(input("Insira os valores da lista:"))
        listaN.append(num)
        N = N - 1
    return listaN

def soma (lista):
    soma = 0
    for i in lista:
        soma = soma + i
    return soma

def media (lista):
    soma = 0
    for i in lista:
        soma = soma + i
    return soma/len(lista)

def maior (lista):
    maior = 0
    for i in lista:
        if i > maior:
            maior = i
    return maior

def menor (lista):
    menor = lista[0]
    for i in lista:
        if i < menor:
            menor = i
    return menor

def crescente (lista):
    i = 1
    while (i < len(lista)):
        if lista[i] < lista[i-1]:
            valor = False
        else:
            valor = True
        i = i + 1
    return valor

def decrescente (lista):
    i = 1
    while (i < len(lista)):
        if lista[i] > lista[i-1]:
            valor = False
        else:
            valor = True
        i = i + 1
    return valor
            
def indice (lista,elem):
    valor = 0
    i = 0
    if elem in lista:
        while (i < len(lista)):
            if lista[i] == elem:
                valor = i
            i = i + 1
    else:
        valor = - 1
    return valor

menu() #mostra o menu todo
opção = int(input("Insira qual a opção que deseja:"))
listaF = [] #a lista começa vazia

while opção != 0:

    if opção == 1:
        listaF.clear() #limpa os dados anteriores
        listaF = criaListaR(int(input("Insira o número de elementos da lista:")))

    if opção == 2:
        listaF.clear()
        listaF = criaLista(int(input("Insira o número de elementos da lista:")))

    if opção == 3:
        print ("A soma dos elementos da lista é:",soma(listaF))

    if opção == 4:
        print ("A média dos elementos da lista é:",media(listaF))

    if opção == 5:
        print ("O maior valor da lista é", maior(listaF))
    
    if opção == 6:
        print ("O menor valor da lista é", menor(listaF))

    if opção == 7:
        if crescente(listaF) == True:
            print ("Sim")
        else:
            print ("Não")
    if opção == 8:
        if decrescente(listaF) == True:
            print ("Sim")
        else:
            print ("Não")
    if opção == 9:
        novoindice = indice(listaF,int(input("Insira o valor do elemento")))
        print (novoindice)
        if novoindice == - 1:
            print ("O elemento não existe na lista")

    opção = int(input ("O que pretende fazer?")) # pergunta no fim de cada opção para ver se queremos sair ou fazer mais algum cálculo

print ("A sua lista é:",listaF)
```
